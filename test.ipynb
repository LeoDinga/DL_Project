```python
from ultralytics import YOLO

model = YOLO('yolo11n-pose.pt')

results = model.predict('p48_serslice_s1 .avi', save=True, save_txt=True )
```

```python
import numpy as np

points_arr=[]
for r in results:
    #print(r)
    k_points = r.keypoints.data.tolist() #data dá: (x, y, conf)
    points_arr.append(k_points)

# print(len(points_arr)) # verificar que temos os keypoints dos frames todos
# print(points_arr) #array com os keypoints por frames

# Guardar os keypoints num ficheiro txt
with open('keypoints.txt', 'w') as f:
    for frame_points in points_arr:
        for point in frame_points:
            f.write(' '.join(map(str, point)) + '\n')
        f.write('\n')  # Linha em branco para separar frames
```
